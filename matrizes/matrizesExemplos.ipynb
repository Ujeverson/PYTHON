```python
import numpy as np
```

## Matriz linha

```python
A = np.array([[1, -3, 6, 8]])
print("A=", A, "\nOrdem:",np.shape(A))
```

## Matriz coluna

```python
B = np.array([[1],[-3],[6],[8]])
print("B=", B, "\nOrdem:",np.shape(B))
```

## Matriz quadrada

```python
C = np.array([[1, 0.1, -2.3, 0.2],[-3, 0, 3 , 4],[6, -7, 81, 2],[8, 7, 6 ,5]])
print("C=\n", C, "\nOrdem:",np.shape(C))
```
